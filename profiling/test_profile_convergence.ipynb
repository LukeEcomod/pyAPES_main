# Plot difference between An, Rd, fe, gs_opt, ci, cs

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
results_high = xr.open_dataset('../results/testrun_test_june.nc')
```

```python
results_mid = xr.open_dataset('../results/testrun_base_june.nc')
```

```python
target_variables = ['pt_net_co2_sunlit','pt_net_co2_shaded',
                    'pt_dark_respiration_sunlit', 'pt_dark_respiration_shaded',
                    'pt_stomatal_conductance_h2o_sunlit','pt_stomatal_conductance_h2o_shaded',
                    'pt_transpiration_sunlit','pt_transpiration_shaded',
                    'pt_leaf_internal_co2_sunlit','pt_leaf_internal_co2_shaded',
                    'pt_leaf_surface_co2_sunlit','pt_leaf_surface_co2_shaded','canopy_lad',
                    'pt_boundary_conductance_h2o_sunlit','pt_boundary_conductance_h2o_shaded',
                    'ffloor_surface_temperature','gt_temperature',
                    'canopy_SWnet','canopy_leaf_net_SW']
```

```python
difference = np.abs(results_high[target_variables] - results_mid[target_variables])/results_high[target_variables]*100
```

```python
for key, value in difference.items():
    distribution = value.values.flatten()
    print(key)
    print(np.nanmin(distribution), np.nanmean(distribution), np.nanmedian(distribution), np.nanmax(distribution))

results_high.close()
results_mid.close()
```

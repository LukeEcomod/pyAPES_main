```python
import pstats
from pstats import SortKey
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
p_simple_soil = pstats.Stats(os.getenv('SIMPLE_SOIL_PROFILE_RUN'))
p_full_soil = pstats.Stats(os.getenv('FULL_SOIL_PROFILE_RUN'))
p_full_soil_moss_iteration = pstats.Stats(os.getenv('FULL_SOIL_MOSS_ITERATION_PROFILE_RUN'))
p_hackathon_day1 = pstats.Stats(os.getenv('HACKATHON_DAY1_PROFILE'))
p_hackathon_day2 = pstats.Stats(os.getenv('HACKATHON_DAY2_PROFILE'))
p_hackathon_day3 = pstats.Stats(os.getenv('HACKATHON_DAY3_PROFILE'))
```

```python
p_simple_soil.sort_stats(SortKey.TIME).print_stats(30)
```
